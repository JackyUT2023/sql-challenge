```python
# Dependencies
import pandas as pd
from pathlib import Path
# Save path to the data set as a variable.
departments_file = Path("EmployeeSQL/Starter_Code/data/departments.csv")
dept_emp_file = Path("EmployeeSQL/Starter_Code/data/dept_emp.csv")
dept_manager_file = Path("EmployeeSQL/Starter_Code/data/dept_manager.csv")
employees_file = Path("EmployeeSQL/Starter_Code/data/employees.csv")
salaries_file = Path("EmployeeSQL/Starter_Code/data/salaries.csv")
titles_file = Path("EmployeeSQL/Starter_Code/data/titles.csv")
```

```python
# Use Pandas to the read data.
departments_df = pd.read_csv(departments_file)
departments_df.head()
```

```python
# Use Pandas to the read data.
dept_emp_df = pd.read_csv(dept_emp_file)
dept_emp_df.head()
```

```python
# Use Pandas to the read data.
dept_manager_df = pd.read_csv(dept_manager_file)
dept_manager_df.head()
```

```python
# Use Pandas to the read data.
employees_df = pd.read_csv(employees_file)
employees_df.head()
```

```python
# Use Pandas to the read data.
salaries_df = pd.read_csv(salaries_file)
salaries_df.head()
```

```python
# Use Pandas to the read data.
titles_df = pd.read_csv(titles_file)
titles_df.head()
```

```python
# List the employee number, last name, first name, sex, and salary of each employee
employee_salary_merge_df = pd.merge (employees_df,salaries_df,on=['emp_no','emp_no'])
renamed_employee_salary_merge_df = employee_salary_merge_df.rename(columns={'emp_no':'employee_number'})
renamed_employee_salary_merge_df[['employee_number','last_name','first_name','sex','salary']].head()
```

```python
# List the first name, last name, and hire date for the employees who were hired in 1986

# Extract the "year" value from "hire_date" column and make a new column.
employees_df ['year'] = pd.DatetimeIndex(employees_df['hire_date']).year
# Use loc function to lock all rows in Year =1986.
employees_df_loc=employees_df.loc[employees_df['year']==1986,:]
# Show columns that were asking
employees_df_loc[['first_name','last_name','hire_date']].head()
```

```python
# List the manager of each department 
# along with their department number, department name, employee number, last name, and first name

emp_dept_manager_merge_df = pd.merge (employees_df,dept_manager_df,on=['emp_no','emp_no'])
emp_dept_manager_merge_df = pd.merge (departments_df,emp_dept_manager_merge_df,on=['dept_no','dept_no'])

rename_emp_dept_manager_merge_df = emp_dept_manager_merge_df.rename(
    columns={'emp_no':'employee_number',
             'dept_no':'department_number',
             'dept_name':'department_name'}
)

rename_emp_dept_manager_merge_df[['employee_number',
                                  'last_name',
                                  'first_name',
                                  'department_number',
                                  'department_name']]
```

```python
# List the department number for each employee 
# along with that employee’s employee number, last name, first name, and department name

emp_dept_merge_df = pd.merge (employees_df,dept_emp_df,on=['emp_no','emp_no'])
emp_dept_merge_df = pd.merge (departments_df,emp_dept_merge_df,on=['dept_no','dept_no'])

rename_emp_dept_merge_df = emp_dept_merge_df.rename(
    columns={'emp_no':'employee_number',
             'dept_no':'department_number',
             'dept_name':'department_name'}
)

rename_emp_dept_merge_df[['employee_number',
                          'last_name',
                          'first_name',
                          'department_name']]
```

```python
# List first name, last name, and sex of each employee 
# whose first name is Hercules and whose last name begins with the letter B

# Create masks
employees_df['last_name_B'] = employees_df['last_name'].str.startswith ('B')
# Loc to filter
employees_df_loc_1=employees_df.loc[employees_df['first_name']=='Hercules',:]
employees_df_loc_2=employees_df_loc_1.loc[employees_df_loc_1['last_name_B']==True,:]
# Show result 
employees_df_loc_2[['first_name','last_name','sex']]
```

```python
# List each employee in the Sales department, including their employee number, last name, and first name
emp_dept_merge_df = pd.merge (employees_df,dept_emp_df,on=['emp_no','emp_no'])
emp_dept_merge_df = pd.merge (departments_df,emp_dept_merge_df,on=['dept_no','dept_no'])

rename_emp_dept_merge_df = emp_dept_merge_df.rename(
    columns={'emp_no':'employee_number',
             'dept_no':'department_number',
             'dept_name':'department_name'}
)

rename_emp_dept_merge_df = rename_emp_dept_merge_df.loc[
    rename_emp_dept_merge_df['department_name']=='Sales'] 

rename_emp_dept_merge_df[['employee_number',
                          'last_name',
                          'first_name',
                          'department_name']]
```

```python
# List each employee in the Sales and Development departments, 
# including their employee number, last name, first name, and department name

emp_dept_merge_df = pd.merge (employees_df,dept_emp_df,on=['emp_no','emp_no'])
emp_dept_merge_df = pd.merge (departments_df,emp_dept_merge_df,on=['dept_no','dept_no'])

rename_emp_dept_merge_df = emp_dept_merge_df.rename(
    columns={'emp_no':'employee_number',
             'dept_no':'department_number',
             'dept_name':'department_name'}
)

rename_emp_dept_merge_df = rename_emp_dept_merge_df.loc[
    rename_emp_dept_merge_df['department_name'].isin (['Sales','Development'])]

rename_emp_dept_merge_df[['employee_number',
                          'last_name',
                          'first_name',
                          'department_name']]
```
